# 96. Unique Binary Search Trees (Medium)

<div><p>Given <em>n</em>, how many structurally unique <strong>BST's</strong> (binary search trees) that store values 1 ...&nbsp;<em>n</em>?</p>

<p><strong>Example:</strong></p>

<pre><strong>Input:</strong> 3
<strong>Output:</strong> 5
<strong>Explanation:
</strong>Given <em>n</em> = 3, there are a total of 5 unique BST's:

   1         3     3      2      1
    \       /     /      / \      \
     3     2     1      1   3      2
    /     /       \                 \
   2     1         2                 3
</pre>
</div>

## Option 1

<p>DP
    <p>
<li>Suppose x is a number between 1...n. We make x the head, the number of different BST's that can be made with x as head is equal to:<br>  k(x-1)*k(n-x)  i.e. the answer to the number of items on the left * the right
<li>This can be proven with a lower number, the table below shows how the result can be scaled up using DP
    <li>The cells show the number of BST's that can be made given the column number as the head node
<p>


    
   <p>
Time complexity: O(n^2)
    <br>
Space complexity: O(n)


```python
class Solution(object):
    def numTrees(self, n):        
        k = [0]*(n+1)
        k[0] = 1
        for i in range(1,n+1):
            for j in range(i-1,-1,-1):
                k[i] += k[j]*k[-(j+1)%i]
        return k[-1]
        
        
Solution().numTrees(8)
```

#### Result: 16ms (70.28%)
