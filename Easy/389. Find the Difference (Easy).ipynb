# 389. Find the Difference (Easy)

<div><p>
Given two strings <b><i>s</i></b> and <b><i>t</i></b> which consist of only lowercase letters.</p>

<p>String <b><i>t</i></b> is generated by random shuffling string <b><i>s</i></b> and then add one more letter at a random position.</p>

<p>Find the letter that was added in <b><i>t</i></b>.</p>

<p><b>Example:</b>
</p><pre>Input:
s = "abcd"
t = "abcde"

Output:
e

Explanation:
'e' is the letter that was added.
</pre></div>

## Option 1

Time complexity = O(t)
<br>
Space complexity = O(s)

```python
import collections
class Solution(object):
    def findTheDifference(self, s, t):
        s_map = collections.Counter(s)
        for i in t:
            if s_map[i] > 0:
                s_map[i] -= 1
            else:
                return i

Solution().findTheDifference("aabb", "aabbb")
```

#### Result: 24ms (65.45%)

## Option 2

Time complexity = O(s)
<br>
Space complexity = O(1)

```python
import collections
class Solution(object):
    def findTheDifference(self, s, t):
        n = 0
        for i in range(len(s)):
            n += ord(t[i])
            n -= ord(s[i])
        n += ord(t[-1])
        return chr(n)
        
Solution().findTheDifference("a", "aa")
```

#### Result: 20ms (86.18%)
